```
import yfinance as yf
tickerSymbol = 'SBIN.NS'
tickerData = yf.Ticker(tickerSymbol)
SBIN = tickerData.history(period='1d', start='2014-04-1', end='2020-03-31')
SBIN
```

```
import yfinance as yf
tickerSymbol = 'HDFC.NS'
tickerData = yf.Ticker(tickerSymbol)
HDFC = tickerData.history(period='1d', start='2014-04-1', end='2020-03-31')
HDFC
```

```
print(SBIN.tail())
print(HDFC.tail())
```

```
import pandas as pd
data = [SBIN['Close'],HDFC['Close']]
data = pd.DataFrame(data).transpose()
data = pd.DataFrame(data.values)
data = data.rename(columns={0:"SBIN",1 : "HDFC" })
data
```

```
import matplotlib.pyplot as plt
data.plot(figsize=(10, 7))
plt.legend()
plt.title("Adjusted Close Price (31/03/2020) of TOP 2 BANKS ", fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
df = data 
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

```
SBIN['Volume'].plot(figsize=(16,8),label = "SBIN",title='Volume')
HDFC['Volume'].plot(label = "HDFC")
plt.legend();
```

```
SBIN['Total Traded'] = SBIN['Open'] * SBIN['Volume']
HDFC['Total Traded'] = HDFC['Open'] * HDFC['Volume']
SBIN['Total Traded'].plot(figsize=(16,8),label = "SBIN",title='Total Money Traded')
HDFC['Total Traded'].plot(label = "HDFC")
plt.legend()
plt.ylabel('Total Money Traded')
```

```
SBIN['returns'] = SBIN['Close'].pct_change(1)
HDFC['returns'] = HDFC['Close'].pct_change(1)
```

```
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(SBIN['returns'][-30:-1])
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("SBIN last month returns data")
plt.show()
```

```
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(HDFC['returns'][-30:-1])
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("HDFC last month returns data")
plt.show()
```

```
SBIN['returns'].hist(bins=100)
```

```
HDFC['returns'].hist(bins=100)
```

```
SBIN['returns'].hist(bins=50,label='SBIN',figsize = (10,8), alpha= 0.5)
HDFC['returns'].hist(bins=50,label='HDFC', alpha= 0.5)
plt.xlabel('Returns')
plt.legend();
```

```
SBIN['daily_cumulative_return'] = ( 1 + SBIN['returns'] ).cumprod()
HDFC['daily_cumulative_return'] = ( 1 + HDFC['returns'] ).cumprod()
SBIN['daily_cumulative_return'].plot(label='SBIN',figsize = (16,8),title = 'Cumilative Return')
HDFC['daily_cumulative_return'].plot(label='HDFC')
plt.legend();
```

```
"""## SBIN"""

print ("SBIN returns mean = ", SBIN['returns'].mean(),"\n\n",
       "SBIN returns mean * 250 = ",SBIN['returns'].mean()*250,"\n\n",
       "SBIN returns Standard Deviation = ",SBIN['returns'].std(),"\n\n",
       "SBIN returns Standard Deviation * 250 * 0.5 = ",SBIN['returns'].std()*250**0.5)
```

```
"""## HDFC"""

print ("HDFC returns mean = ", HDFC['returns'].mean(),"\n\n",
       "HDFC returns mean * 250 = ",HDFC['returns'].mean()*250,"\n\n",
       "HDFC returns Standard Deviation = ",HDFC['returns'].std(),"\n\n",
       "HDFC returns Standard Deviation * 250 * 0.5 = ",HDFC['returns'].std()*250**0.5)
```
